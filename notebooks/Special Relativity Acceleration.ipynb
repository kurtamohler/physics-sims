### Can special relativity handle accelerations?

This section is a bit long-winded. If you already accept that special relativity can account for accelerations, then skip this.

I have heard claims that special relativity theory doesn't account for accelerating objects. I've heard that general relativity is needed to do it, and that special relativity is the special case of general relativity where objects are moving at constant velocities. But are these claims true? I'll bring up several points to try to address this in an intuitive sense. Then we'll get into the math to see what special relativity can actually say about acceleration.

Firstly, in order to derive the Lorentz transformations, we never had to make the assumption that all objects must remain at a constant speed, so our postulates certainly do not explicitly rule out the possibility of handling acceleration. Remember, we just modified the Galilean relativity postulates to replace the universal time postulate with the constant speed of light postulate, and Galilean relativity certainly can handle accelerations.

So in order to accept the claim that special relativity cannot handle accelerating objects, then it would have to be possible to demonstrate this shortcoming by attempting to handle acceleration and then finding some sort of contradiction or self-inconsistency in the predictions that special relativity provides. This would be a proof by contradiction. I personally have not seen such an example yet, so I don't accept the claim that special relativity cannot handle acceleration.

Secondly, in the limit where the speed of light is infinite, the Lorentz transformation becomes exactly the same thing as the Galilean transformation. So this means that special relativity definitely can at least handle accelerations of objects whose relative velocities remain much less than the speed of light. So at best, we would have to modify the claim that "special relativity cannot handle acceleration" to instead be "special relativity cannot handle accelerations of objects that are moving at high speeds in a given inertial frame, but the theory can handle accelerations of objects moving at speeds much less than the speed of light in a given inertial frame" in order to have any chance of the claim being true.

Thirdly, the claim that special relativity is the special case of general relativity where objects are moving at constant velocities is evidently not true--at least not based on the descriptions of special and general relativity that I have found, which say that the difference between the two theories is that general relativity accounts for gravity and special relativity does not. Gravity is the specific difference between the two theories. Supposedly, in the limit where the strength of the gravitational field (not really a field, but still "field" seems to be used colloquially) goes to 0, general relativity reduces to special relativity.

Fourthly, electrodynamics and classical field theory take place within a spacetime which exhibits Lorentz invariance. Special relativity is the arena in which classical fields exist. Fields apply accelerations to particles. If electrodynamics and classical field theory are consistent with special relativity, and if objects can accelerate in those theories, then special relativity does handle accelerations.

Fifthly, many of the thought experiments in special relativity that people first encounter actually do contain accelerations. Take the twin paradox, for instance. Alice and Bob are twins of the same age. Alice takes a trip on a spaceship to some distant star at a high velocity $v$ and then turns around to arrive back at Earth. How did Alice achieve her high velocity? How did she turn around? The answer is that she accelerated. In the usual formulation of this problem, she changes velocity instantly. At one instant, she is at rest with respect to Earth, and in the next instant, she is travelling at speed $v$. This can be thought of as a discrete acceleration, where velocity changes instantly by a finite amount. So special relativity can at least handle discrete acceleration. (Note that the name "twin paradox" is a misnomer. It's not actually a paradox, and it only seems like a paradox if you forget that simultaneity is relative. But I won't say more about the twin paradox here.)

Sixthly, a continuous acceleration can actually be constructed out of infinitely many discrete accelerations. If we have an object that is moving at a constant velocity of $v=v_{start}=0$ at time $t=t_{start}=0$, we can discretely accelerate it at time $t=\Delta t$ by changing its velocity to $v=\Delta v$. Then at time $t=2\Delta t$, we can change the velocity again to be $v=2\Delta v$. We can continue doing that for $n$ steps, so that at time $t=n\Delta t=t_{end}$, the velocity is changed to $v=n\Delta v=v_{end}$. We just have to make sure that $|v_{end}| < c$.

The function of velocity over time is a staircase. Averaging between time $t=0$ and $t=t_{end}$, the object experiences an average acceleration of $a = \dfrac{v_{end}-v_{start}}{t_{end}-t_{start}} = \dfrac{n\Delta v - 0}{n\Delta t - 0}= \dfrac{\Delta v}{\Delta t}$.

Now we can shrink the sizes of $\Delta t$ and $\Delta v$ to be infinitessimally small and at the same time increase the number of steps $n$ to infinity, while keeping $t_{end}$, $v_{end}$, and $a$ fixed. Then we can say that the object is experiencing a continuous acceleration of $a = \dfrac{dv}{dt}$ between the times $t=t_{start}$ and $t=t_{end}$. So if special relativity allows for discrete accelerations, it seems like it must be able to account for continuous accelerations as well, since we can construct continuous accelerations from discrete ones.

Now it's time to actually go ahead and look into how accelerations would transform, if in fact they can exist within the theory.

### Deriving the Lorentz acceleration transformation

Let's say we have some particle that has position $x$, velocity $\dot{x}$, and acceleration $\ddot{x}$ at time $t$ in frame $S$. We want to transform these to another frame $S'$ that is moving with velocity $v$ with respect to $S$. We already know how to transform the time and position coordinate with the Lorentz transformation:

$$ t'=\frac{t-vx/c^2}{\sqrt{1-(v/c)^2}} $$
$$ x'=\frac{x-vt}{\sqrt{1-(v/c)^2}} $$

We also know how to transform the velocity with the Lorentz velocity transformation:

$$ \dot{x}' = \frac{\dot{x} - v}{1 - v\dot{x}/c^2} $$

Now we have to derive the acceleration transformation. We can do it using essentailly the same strategy that we used to derive the velocity transformation equation. For that, we took the time derivative of the Lorentz transformation equation for $x'$ to find an equation for $\dot{x}'$. Now, we'll take the time derivative of the equation for $\dot{x}'$ to find an equation for $\ddot{x}'$.

$$ \ddot{x}' = \frac{d}{dt'}\dot{x}' = \frac{d}{dt'}\frac{\dot{x} - v}{1 - v\dot{x}/c^2} $$

We have to use the product rule for this.

$$ \ddot{x}' = \frac{d}{dt'}AB = A\frac{dB}{dt'} + B\frac{dA}{dt'} $$
$$ A=\dot{x}-v $$
$$ B = \frac{1}{1-v\dot{x}/c^2} $$
$$ \frac{dA}{dt'}=\frac{d\dot{x}}{dt'} $$

To find $\dfrac{dB}{dt'}$, we have to use the chain rule.

$$ \frac{dB}{dt'} = \frac{dB}{dC}\frac{dC}{dt'} $$
$$ B = \frac{1}{1 - v\dot{x}/c^2} = \frac{1}{C} = C^{-1}$$
$$ C = 1 - v\dot{x}/c^2 = 1 - \frac{v}{c^2}\dot{x} $$
$$ \frac{dB}{dC} = -C^{-2} = -\frac{1}{C^2} = -\frac{1}{(1 - v\dot{x}/c^2)^2} $$
$$ \frac{dC}{dt'} = -\frac{v}{c^2}\frac{d\dot{x}}{dt'} $$
$$ \frac{dB}{dt'} = \frac{v}{c^2(1 - v\dot{x}/c^2)^2}\frac{d\dot{x}}{dt'} $$

We can now plug everything into the product rule formula.

$$ \ddot{x}' = A\frac{dB}{dt'} + B\frac{dA}{dt'} = (\dot{x} - v) (\frac{v}{c^2(1 - v\dot{x}/c^2)^2}\dfrac{d\dot{x}}{dt'}) + (\frac{1}{1 - v\dot{x}/c^2})(\frac{d\dot{x}}{dt'})$$
$$ \ddot{x}' = (\frac{v\dot{x} - v^2}{c^2(1 - v\dot{x}/c^2)^2} + \frac{1}{1 - v\dot{x}/c^2}) \frac{d\dot{x}}{dt'}$$
$$ \ddot{x}' = \frac{v\dot{x} - v^2 + c^2(1 - v\dot{x}/c^2)}{c^2(1 - v\dot{x}/c^2)^2} \frac{d\dot{x}}{dt'}$$
$$ \ddot{x}' = \frac{v\dot{x} - v^2 + c^2 - v\dot{x}}{c^2(1 - v\dot{x}/c^2)^2} \frac{d\dot{x}}{dt'}$$
$$ \ddot{x}' = \frac{c^2 - v^2}{c^2(1 - v\dot{x}/c^2)^2} \frac{d\dot{x}}{dt'}$$

To make sense of $\dfrac{d\dot{x}}{dt'}$, we can use the chain rule:

$$ \frac{d\dot{x}}{dt'} = \frac{d\dot{x}}{dt}\frac{dt}{dt'} = \ddot{x} \frac{dt}{dt'}$$

We found $\dfrac{dt}{dt'}$ in the [main notes on special relativity](./Special%20Relativity.ipynb), so we'll reuse it here.

$$ \frac{dt}{dt'} = \frac{\sqrt{1 - (v/c)^2}}{1 - v\dot{x}/c^2} $$
$$ \frac{d\dot{x}}{dt'} = \ddot{x} \frac{\sqrt{1 - (v/c)^2}}{1 - v\dot{x}/c^2}$$
$$ \ddot{x}' = \frac{c^2 - v^2}{c^2(1 - v\dot{x}/c^2)^2} \ddot{x} \frac{\sqrt{1 - (v/c)^2}}{1 - v\dot{x}/c^2}$$
$$ \ddot{x}' = \frac{(c^2 - v^2)\sqrt{1 - (v/c)^2}}{c^2(1 - v\dot{x}/c^2)^3} \ddot{x}$$
$$ \ddot{x}' = \frac{(1 - (v/c)^2)\sqrt{1 - (v/c)^2}}{(1 - v\dot{x}/c^2)^3} \ddot{x}$$
$$ \ddot{x}' = \frac{(1 - (v/c)^2)^{3/2}}{(1 - v\dot{x}/c^2)^3} \ddot{x}$$

And there we have the Lorentz acceleration transformation for one spatial dimension.

This is almost the same equation for acceleration transformation given in "Classical Electrodynamics" by Jackson. Although in that book, the equation is generalized to accelerations in three spatial dimensions, and it transforms from the $S'$ frame to the $S$ frame, whereas ours goes the other way. Below is an excerpt from the book.

<img src="images/jackson-acceleration-transformation.png" />

Note that in the limit as $c$ goes to infinity, the acceleration has exactly the same value in both frames. That is what we would expect from the Galilean relativity, because if the laws of physics are the same in every frame and time is universal, then accelerations have to be the same in every frame.

### Acceleration in inertial frames cannot be constant

Remember that the speed $|\dot{x}|$ of an object in any inertial reference frame $S$ is limited to the speed of light $c$. Because of that, in any inertial frame $S$, an object cannot maintain a constant acceleration $\ddot{x}=g$ forever. This is a very important quality about accelerations in special relativity, and it is why the acceleration transformation equation depends on $\dot{x}$.

An object could in principle maintain a constant acceleration for a finite amount of time, but at some point, the magnitude of the acceleration in the inertial frame would have to decrease before the speed reaches $c$.

Another important observation is that $\ddot x$ at a particular time does not appear to be bounded. It could be an incredibly large number, just as long as it decreases later on in time, to keep the velocity within the speed limit.

### Example: Transforming a particle's acceleration

Going back to the example of a particle at a particular time $t$ with position $x$, velocity $\dot{x}$, and acceleration $\ddot{x}$, let's consider some concrete examples and observe what happens in different reference frames.

Let's say the particle's acceleration is $\ddot{x} = 1$ and its velocity is $\dot{x}=0$ at some particular time in inertial frame $S$. Let's calculate the transformed acceleration and velocity for another inertial frame $S'$. We'll calculate the answer for the entire range of possible transformation velocities. We'll use $c=1$.

```python
import numpy as np
import matplotlib.pyplot as plt

x_ddot = 1
x_dot = 0
c = 1
v_transform = np.linspace(-c, c, 10_000)

def transform_x_ddot(x_ddot, x_dot, v, c):
    return x_ddot * ((1 - (v/c)**2)**(0.5) / (1 - v * x_dot / c**2))**3

def transform_x_dot(x_dot, v, c):
    return (x_dot - v) / (1 - v * x_dot / c**2)

x_ddot_prime = transform_x_ddot(x_ddot, x_dot, v_transform, c)
x_dot_prime = transform_x_dot(x_dot, v_transform, c)

plt.plot(v_transform, x_ddot_prime, label="acceleration ẍ'")
plt.plot(v_transform, x_dot_prime, label="velocity ẋ'")

plt.title(f"ẍ={x_ddot}, ẋ={x_dot} transformed to other reference frames")
plt.xlabel("v, velocity of frame S' with respect to S")
plt.ylabel("transformed ẍ' or ẋ'")
plt.legend()
plt.grid()
plt.show()
```

In the limit as the transformation velocity $v$ approaches $c=\pm 1$, we can see that the acceleration goes to 0.

An intuitive explanation for why this happens is time dilation. If we put a clock at rest in frame $S$, and if frame $S'$ approaches velocity $v=\pm c$, then the frequency of the clock's ticking in frame $S'$ approaches 0. So the process of the acceleration acting on the particle at rest in frame $S$ will get stretched out through time $t'$ in frame $S'$. In other words, it will take longer for the particle's speed in frame $S'$ to change because of the time dilation. This makes the acceleration $\ddot{x}'$ in frame $S'$ decrease towards 0 as the transformation velocity approaches $v=\pm c$.

Let's also see what happens if the particle has a nonzero velocity in frame $S$. I'll use $\dot{x}=0.5$ for this one.

```python
x_ddot = 1
x_dot = 0.5

x_ddot_prime = transform_x_ddot(x_ddot, x_dot, v_transform, c)
x_dot_prime = transform_x_dot(x_dot, v_transform, c)

plt.plot(v_transform, x_ddot_prime, label="acceleration ẍ'")
plt.plot(v_transform, x_dot_prime, label="velocity ẋ'")

plt.title(f"ẍ={x_ddot}, ẋ={x_dot} transformed to other reference frames")
plt.xlabel("v, velocity of frame S' with respect to S")
plt.ylabel("transformed ẍ' or ẋ'")
plt.legend()
plt.grid()
plt.show()
```

Here we see something similar to the last plot, but it looks as if the bell curve of the acceleration from the previous plot is sort of stretched toward the right so that the maximum transformed acceleration $\ddot{x}'$ happens at $v=0.5$. Why is that where the maximum is? Well, that's because we set the particle's velocity in $S$ to be $\dot{x}=0.5$, so the frame in which the particle is at rest is the frame $S'$ that is moving at velocity $v=\dot{x}=0.5$ with respect to $S$. As we saw in the previous example, the frame where the particle is at rest is the frame where the acceleration is the greatest. And we can confirm in this plot that the maximum acceleration also occurs when $\dot{x}' = 0$.

### Example: Accelerating spaceship

Now let's look at a specific problem.

Imagine a spaceship that is accelerating such that a person inside of it feels a constant acceleration $g$. Can we find a function $x=\phi(t)$ that describes the spaceship's motion in some inertial reference frame $S$?

(This is problem 11.5 in the book "Classical Electrodynamics" by Jackson.)

To begin this problem, I think it's instructive to consider what happens in the non-relativistic limit. That is, consider a section of the spaceship's worldline where its speed $|\dot{x}|$ in the inertial frame $S$ is much less than the speed of light. During that section of the worldline, a clock in the spaceship will tick at approximately the same (or negligibly different) rate as a clock that is at rest in $S$, and because of that, the acceleration $\ddot{x}$ of the spaceship in frame $S$ will be approximately the same as the acceleration that the person in the spaceship feels, so $\ddot{x}=g$ for that section.

A curve defined by $\ddot{x}=g$ is just a parabola, which we can find by taking the time integral twice. I'll choose the origin of $S$ to be located at the stationary point of the parabola, so that the path is described by $x=\frac{1}{2}gt^2$. Note that $\dot{x}=gt$ and $\ddot{x}=g$, satisfying the correct acceleration. Below is a plot of this worldline.

<!-- Google Drawings: https://docs.google.com/drawings/d/1plYFckWtHOPLCfW9Mje7RQ3wPonyIud1MxhKCmnZaIU/edit?usp=sharing -->
<img src="images/sr-acceleration-simple.png" />

However, the real answer must account for special relativity. According to the parabolic equation, the velocity of the spaceship is $\dot{x}=gt$, which is unbounded--as time increases toward infinity, the speed increases toward infinity as well. But we know that the spaceship cannot travel faster than light, $|\dot{x}| < c$. So evidently, the actual path of the spaceship must be shaped similarly to a parabola, except that its arms are stretched out much further, so that as time increases toward infinity, the speed is bound to be less than $c$.

What can we say about the frame of reference of the accelerating spaceship? There's a key observation that we can make which will be very helpful: at each moment in time $t$, the spaceship is moving at some definite velocity $\dot x=\dot \phi (t)$, and we can consider an inertial reference frame $S'$ which is moving at the same velocity as the spaceship at that time. We can call such an inertial frame the "instantaneous" inertial frame $S'$ in which the spaceship is at rest at a certain time $t$. In this case, $S'$ does not represent one single reference frame, it represents an infinite number of different instantaneous frames, each at a different time $t$.

<!-- Google Drawings: https://docs.google.com/drawings/d/13yYVCBWpUjzGd2RV3P8ePAnLErEudffYEpj3QY035Ng/edit?usp=sharing -->
<img src="images/sr-acceleration-simple-instantaneous-frames.png" />

Above is a plot containing one such instantaneous frame. I have labeled an event $A$ along the spaceship's path of motion. The 4-position of event $A$ is $(t_A, x_A)$. The velocity of the spaceship is $\dot x = \dot{\phi}(t_A)$ at that event, so the velocity of the instantaneous frame $S'$ is also $v=\dot{x}=\dot \phi (t_A)$. Event $A$ is not special--we'll have the same relationship for all frames $S'$ for each event along the spaceship's path, so in general $v=\dot{x}=\dot \phi (t)$.

Since the frame $S'$ moves at the same velocity as the spaceship at event $A$, then the spaceship is moving at velocity $\dot{x}'=0$ in $S'$. Because of that, the instantaneous acceleration $\ddot{x}'$ of the spaceship in frame $S'$ at event $A$ must be equal to the acceleration $g$ of the spaceship in its own non-inertial frame. So we have $\ddot x' = g$.

To recap, we have found the following relationships:

$$ v = \dot{x} = \dot \phi (t) $$
$$ \ddot{x}'=g $$

Now we can just use the Lorentz acceleration transformation that we derived earlier to find $\ddot{x}$.

$$ \ddot{x}' = \frac{(1 - (v/c)^2)^{3/2}}{(1 - v\dot{x}/c^2)^3} \ddot{x}$$
$$ g = \frac{(1 - (\dot{x}/c)^2)^{3/2}}{(1 - (\dot{x}/c)^2)^3} \ddot{x}$$
$$ g = \frac{1}{(1 - (\dot{x}/c)^2)^{3/2}} \ddot{x}$$
$$ \ddot{x} = g{(1 - (\dot{x}/c)^2)^{3/2}} $$

We've found an equation of motion that describes the path of the spaceship in the inertial reference frame $S$. It's a second order differential equation. It might be possible to analytically integrate it twice over time to get an algebraic equation for $x$. But I'm eager to just numerically integrate it because it's easier.

I will use the initial conditions $(t_0,x_0)=(0, 0)$ and $\dot x (t_0)= 0$. For the constants, I'll use $c=1$ and $g=0.2$.

```python
import physics_sims

v0 = 0
event0 = [0, 0]
g = 0.2
c = 1

class ConstantProperAcceleration(physics_sims.Sim):
    def __init__(self, event0, v0, g, c=1):
        self.event = event0
        self.v = v0
        self.g = g
        self.a = g
        self.c = c

    def calc_a(self, t, x, v):
        res = self.g * (1 - (v/self.c)**2)**(3/2)
        return res
    
    def update(self, sim_runner, dt):
        self.event[0], self.event[1], self.v = physics_sims.integrators.runge_kutta_4th_order(
            dt, self.event[0], self.event[1], self.v,
            self.calc_a
        )
        self.a = self.calc_a(self.event[0], self.event[1], self.v)
    
    def state(self):
        return [self.event[0], self.event[1], self.v, self.a]

traj = physics_sims.SimRunner().run_headless(
    ConstantProperAcceleration(event0, v0, g),
    20,
    time_delta=0.01)

fig, axes = plt.subplots(1, 3)
fig.set_figwidth(15)

plt.suptitle('Frame S')

axes[0].plot(traj.T[1], traj.T[0])
axes[0].set(xlabel='x, position', ylabel='t')
axes[0].axis('equal')

axes[1].plot(traj.T[2], traj.T[0])
axes[1].set(xlabel='x_dot, velocity', ylabel='t')

axes[2].plot(traj.T[3], traj.T[0])
axes[2].set(xlabel='x_ddot, acceleration', ylabel='t')

for ax in axes:
    ax.grid()

plt.show()
```

Above are plots of the position, velocity, and acceleration of the spaceship in frame $S$ over time. Near $t=0$, the acceleration is at its maximum, and the velocity is increasing approximately linearly. As time goes on, the acceleration decreases and velocity increases--similarly to what we saw in the previous example. While the acceleration decreases, the velocity starts changing more slowly. Eventually, the velocity gets so close to the speed limit that the acceleration effectively becomes 0, so the velocity stays just below $c=1$.

The position of the spaceship increases similarly to a parabola in the beginning, but as time goes on, the motion becomes closer and closer to a straight line with velocity just less than $\dot{x}=c$. So it matches what we intuitively guessed the path would look like earlier.

### Varying the acceleration over time

Now I want to see what happens if the acceleration of the spaceship changes over time in its own frame. Here's problem 11.5 from Jackson's "Classical Electrodynamics":

<img src="images/jackson-accelerating-spaceship.png" />

We'll accelerate the spaceship in one direction for 5 years measured by a clock that is in the spaceship, then accelerate in the other direction for 10 years, and then accelerate in the original direction for another 5 years. To do that, we'll have to keep track of the proper time $\tau$ that passes in the spaceship's non-inertial frame. The proper time will start at $\tau=t=0$. We'll have to integrate $\dfrac{dt'}{dt}$ to find the proper time. Remember from before we had:

$$ \frac{dt}{dt'} = \frac{\sqrt{1 - (v/c)^2}}{1 - v\dot{x}/c^2} $$
$$ v = \dot{x} $$
$$ \frac{dt}{dt'} = \frac{\sqrt{1 - (\dot{x}/c)^2}}{1 - (\dot{x}/c)^2} $$
$$ \frac{dt'}{dt} = \sqrt{1 - (\dot{x}/c)^2} $$
$$ \tau = \int_0^t \frac{dt'}{dt} dt $$
$$ \tau = \int_0^t \sqrt{1 - (\dot{x}/c)^2} dt $$

Since we don't have an exact expression for $\dot{x}$, we'll just numerically integrate $/tau$ in the simulation. To make the spaceship comfortable for the astronauts, we'll use an acceleration magnitude of $g=9.8$ $m/s^2$. Let's convert that to units of $\text{light-years}/\text{year}^2$.

$$ g=9.8 \frac{m}{s^2} \cdot \frac{1}{300,000,000} \frac{\text{light-years}}{m} \cdot (60\cdot60\cdot24\cdot365\frac{s}{year})^2 $$

```python
9.8 * 60 * 60 * 24 * 365 / 300_000_000
```

$$ g=1.03 \text{ light-years}/\text{year}^2 $$

```python
v0 = 0
event0 = [0, 0]
g = 1.03
c = 1

class TwinParadoxAcceleration(ConstantProperAcceleration):
    def __init__(self, event0, v0, g, c=1):
        super().__init__(event0, v0, g, c)
        self.proper_time = 0
        self.g0 = g
    
    def calc_a(self, t, x, v):
        if self.proper_time >= 20:
            self.g = 0
        elif self.proper_time >= 0 and self.proper_time < 5:
            self.g = self.g0
        elif self.proper_time >= 5 and self.proper_time < 15:
            self.g = -self.g0
        else:
            self.g = self.g0

        return super().calc_a(t, x, v)
    
    def calc_dt_prime_by_dt(self):
        return np.sqrt(1 - (self.v/self.c)**2)
    
    def update(self, sim_runner, dt):
        _, self.proper_time, _ = physics_sims.integrators.runge_kutta_4th_order(
            dt, self.event[0], self.proper_time, self.calc_dt_prime_by_dt(),
            lambda t, x, v: 0
        )
        super().update(sim_runner, dt)
    
    def state(self):
        if self.proper_time >= 20:
            return None
        else:
            return super().state()

traj = physics_sims.SimRunner().run_headless(
    TwinParadoxAcceleration(event0, v0, g, c),
    350,
    time_delta=0.01)

fig, axes = plt.subplots(1, 3)
fig.set_figwidth(15)

plt.suptitle('Frame S')

axes[0].plot(traj.T[1], traj.T[0])
axes[0].set(xlabel='x, position', ylabel='t')
axes[0].axis('equal')

axes[1].plot(traj.T[2], traj.T[0])
axes[1].set(xlabel='x_dot, velocity', ylabel='t')

axes[2].plot(traj.T[3], traj.T[0])
axes[2].set(xlabel='x_ddot, acceleration', ylabel='t')

for ax in axes:
    ax.grid()

print(f"trip time: {traj[-1][0]} years")
print(f"trip distance: {np.max(traj.T[1])} light-years")

plt.show()
```
